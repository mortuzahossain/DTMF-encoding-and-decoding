```python
import numpy
import math
import matplotlib.pyplot as plt
import wave
import sys
%matplotlib inline 
from scipy import signal
```

```python
def frequencyGeneration(n):
    if n == '1':
        f1 = 697
        f2 = 1209
        return f1,f2
    if n == '2':
        f1 = 697
        f2 = 1336
        return f1,f2
    if n == '3':
        f1 = 697
        f2 = 1477
        return f1,f2
    if n == '4':
        f1 = 770
        f2 = 1209
        return f1,f2
    if n == '5':
        f1 = 770
        f2 = 1336
        return f1,f2
    if n == '6':
        f1 = 770
        f2 = 1477
        return f1,f2
    if n == '7':
        f1 = 852
        f2 = 1209
        return f1,f2
    if n == '8':
        f1 = 852
        f2 = 1336
        return f1,f2
    if n == '9':
        f1 = 852
        f2 = 1477
        return f1,f2
    if n == '0':
        f1 = 941
        f2 = 1336
        return f1,f2
    if n == 'A':
        f1 = 697
        f2 = 1633
        return f1,f2
    if n == 'B':
        f1 = 770
        f2 = 1633
        return f1,f2
    if n == 'C':
        f1 = 852
        f2 = 1633
        return f1,f2
    if n == 'D':
        f1 = 941
        f2 = 1633
        return f1,f2
    if n == '*':
        f1 = 941
        f2 = 1209
        return f1,f2
    if n == '#':
        f1 = 941
        f2 = 1477
        return f1,f2
```

```python
def frequencytodigit(f1,f2):
    print 'Processed Frequency : ',
    print f1,f2
    dilled_digit = ''
    if f1 >= 941 and f2 <= 1336:
        dilled_digit = 0;
    elif f1 >= 941 and f2 <= 1477:
        dilled_digit = '#';
    elif f1 >= 941 and f2 <= 1209:
        dilled_digit = '*';
    elif f1 >= 842 and f2 <= 1209:
        dilled_digit = '7';
    elif f1 >= 842 and f2 <= 1336:
        dilled_digit = '8';
    elif f1 >= 842 and f2 <= 1477:
        dilled_digit = '9';
    elif f1 >= 760 and f2 <= 1209:
        dilled_digit = '4';
    elif f1 >= 760 and f2 <= 1336:
        dilled_digit = '5';
    elif f1 >= 760 and f2 <= 1477:
        dilled_digit = '6';
    elif f1 >= 697 and f2 <= 1209:
        dilled_digit = '1';
    elif f1 >= 697 and f2 <= 1336:
        dilled_digit = '2';
    elif f1 >= 697 and f2 <= 1477:
        dilled_digit = '3';
    elif f1 >= 941 and f2 <= 1633:
        dilled_digit = 'D';
    elif f1 >= 842 and f2 <= 1633:
        dilled_digit = 'C';
    elif f1 >=760  and f2 <= 1633:
        dilled_digit = 'B';
    elif f1 >= 697 and f2 <= 1633:
        dilled_digit = 'A';
        
    print dilled_digit
```

```python
def process(n):
    f1,f2 = frequencyGeneration(n)
    t = list(numpy.arange(0, 0.05, 0.000125))
    fs = 8000
    y = [math.sin(2*3.1416*f1*t[i]) + math.sin(2*3.1416*f2*t[i]) for i in range(len(t))]
    plt.title('Signal Wave...'+str(n))
    plt.plot(y)
    plt.show()


    rmain=2048*2
    rmag=1024*2
    cn=9
    cr=0.5
    cl=.25
    ch=.28

    # First Filter Low Frequency Signals
    b, a = signal.cheby1(cn,cr,cl)
    yfilt1 = signal.filtfilt(b, a, y)
    plt.title('Low Frequency Signals '+str(n))
    plt.plot(yfilt1)
    plt.show()


    # 
    h2=numpy.fft.fft(yfilt1,rmain)
    hmag2=abs(h2[1:rmag])
    [b1,a1]=signal.cheby1(cn,cr,ch,'high')
    yfilt2=signal.filtfilt(b1,a1,y)

    plt.title('High Frequency Signals ' + str(n))
    plt.plot(yfilt2)
    plt.show()

    h3=numpy.fft.fft(yfilt2,rmain)
    hmag3=abs(h3[1:rmag])

    hlow=numpy.fft.fft(yfilt1,rmain);
    hmaglow=abs(hlow);
    plt.title('FFT Low Pass' + str(n))
    plt.plot(hmaglow[1:rmag])
    plt.show()

    # FFT High Pass
    hhigh=numpy.fft.fft(yfilt2,rmain)
    hmaghigh=abs(hhigh)
    plt.title('FFT High Pass' + str(n))
    plt.plot(hmaghigh[1:rmag])
    plt.show()


    m=max(abs(hmag2))
    n=max(abs(hmag3));
    o = numpy.where(m==hmag2)[0]
    p = numpy.where(n==hmag3)[0]

    f1=((o-1)*fs)/rmain;
    f2=((p-1)*fs)/rmain;

    print frequencytodigit(f1,f2)
```

```python
def main():
    dial = raw_input('DTMF INPUT:')
    for i in range(len(dial)):
        process(dial[i])
```
